```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Buat DataFrame contoh
data = {
    'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun'],
    'Penjualan': [100, 150, 130, 180, 200, 170]
}
df = pd.DataFrame(data)

# Tampilkan statistik deskriptif
print("Statistik Deskriptif:")
print(df.describe())

# Visualisasi dengan Matplotlib
plt.figure(figsize=(8, 5))
plt.plot(df['Bulan'], df['Penjualan'], marker='o', linestyle='-', color='b', label='Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.title('Grafik Penjualan per Bulan')
plt.legend()
plt.grid()
plt.show()

# Visualisasi dengan Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Bulan', y='Penjualan', data=df, palette='viridis')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.title('Grafik Batang Penjualan per Bulan')
plt.show()
```
